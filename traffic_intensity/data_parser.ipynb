```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file = 'data/pomiar_natezenia_drogowego.CSV'

df = pd.read_csv(csv_file)

```

```python
'''Visualize data for specific location '''


identifier = 'MQ_K1'
selected_rows = df.query(f'Id == "{identifier}"')

for date in selected_rows['Data'].unique():
    records_for_date = selected_rows[selected_rows['Data'] == date]

    # Convert the 'Godzina' column to datetime
    records_for_date['Godzina'] = pd.to_datetime(records_for_date['Godzina'], format='%H:%M:%S')
    # Extract only the hour from the 'Godzina' column
    records_for_date['Godzina'] = records_for_date['Godzina'].dt.hour

    # Create a plot
    plt.figure(figsize=(10, 6))
    plt.plot(records_for_date['Godzina'], records_for_date['Natenie poj/godz'], marker='o')
    plt.title(f'Traffic Intensity Changes for Identifier {identifier} on {date}')
    plt.xlabel('Hour')
    plt.ylabel('Traffic Intensity (vehicles/hour)')
    plt.grid(True)
    plt.show()
```

```python
'''
Now after visualization we try to normalize (0,1) traffic for every day and hour for specific location

'''
normalized_data = pd.DataFrame()

for date in selected_rows['Data'].unique():
    records_for_date = selected_rows[selected_rows['Data'] == date]

    # Convert the 'Godzina' column to datetime and extract the hour
    records_for_date['Godzina'] = pd.to_datetime(records_for_date['Godzina'], format='%H:%M:%S').dt.hour

    # Normalize 'Natenie poj/godz' column between 0 and 1
    records_for_date['Znormalizowane Natenie'] = (records_for_date['Natenie poj/godz'] - records_for_date['Natenie poj/godz'].min()) / (records_for_date['Natenie poj/godz'].max() - records_for_date['Natenie poj/godz'].min())

    # Extract the day of the week
    records_for_date['Dzien tygodnia'] = pd.to_datetime(records_for_date['Data'], format='%d.%m.%Y').dt.day_name()

    # Append the records to the normalized_data DataFrame
    normalized_data = pd.concat([normalized_data, records_for_date[['Id', 'Dzien tygodnia', 'Godzina', 'Znormalizowane Natenie']]])

# Save the normalized data to a new CSV file
print(normalized_data)
normalized_data.to_csv(f'{identifier}.csv', index=False)
```

```python
'''
Now we gonna do the same but for all location
'''

selected_identifiers = [f'MQ_K{i}_1' for i in range(1,84)]


for identifier in selected_identifiers:

    selected_rows = df.query(f'Id == "{identifier}"')
    normalized_data = pd.DataFrame()

    for date in selected_rows['Data'].unique():
        records_for_date = selected_rows[selected_rows['Data'] == date]

        # Convert the 'Godzina' column to datetime and extract the hour
        records_for_date['Godzina'] = pd.to_datetime(records_for_date['Godzina'], format='%H:%M:%S').dt.hour

        # Normalize 'Natenie poj/godz' column between 0 and 1
        records_for_date['Natenie'] = (records_for_date['Natenie poj/godz'] - records_for_date['Natenie poj/godz'].min()) / (records_for_date['Natenie poj/godz'].max() - records_for_date['Natenie poj/godz'].min())

        # Extract the day of the week
        records_for_date['Dzien tygodnia'] = pd.to_datetime(records_for_date['Data'], format='%d.%m.%Y').dt.day_name()

        # Append the records to the normalized_data DataFrame
        normalized_data = pd.concat([normalized_data, records_for_date[['Id', 'Dzien tygodnia', 'Godzina', 'Natenie']]])
    print(normalized_data)
    if len(normalized_data)>0:
        normalized_data.to_csv(f'data/normalize_traffic/{identifier}.csv', index=False)
```

```python
'''
Now we calculate global avarage traffic for every day and hour
'''
import os
result_df = pd.DataFrame(columns=['Dzien tygodnia', 'Godzina', 'Srednie natenie'])


file_paths = os.listdir('data/normalize_traffic')

for file_path in file_paths:

    df = pd.read_csv(f'data/normalize_traffic/{file_path}')

    grouped_df = df.groupby(['Dzien tygodnia', 'Godzina'])['Natenie'].mean().reset_index()


    result_df = pd.concat([result_df, grouped_df])

result_df['Srednie natenie'] = result_df.groupby('Godzina')['Natenie'].transform('mean')

print(result_df)
result_df.to_csv('global_traffic.csv', index=False)
```

```python
'''
Now lets calculate global avarage traffic for every day and every hour
'''
import csv
result_df = pd.DataFrame(columns=['Dzien tygodnia', 'Godzina', 'Srednie natenie'])

file_paths = os.listdir('data/normalize_traffic')


sum_traffic = [[0,0,0] for i in range(168)]

for file in file_paths:
    df = pd.read_csv(f'data/normalize_traffic/{file}')
    num_rows = df.shape[0]
    for i in range(num_rows):
        data = df.iloc[i].values
        sum_traffic[i][1] = data[1]
        sum_traffic[i][2] = data[2]
        if not pd.isna(data[-1]):
            sum_traffic[i][0] += data[-1]

for row in sum_traffic:
    row[0] /= len(file_paths)

print(sum_traffic)

csv_file = 'data/global_avarage_traffic.csv'


with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Traffic","Week_day", "Hour"])
    writer.writerows(sum_traffic)
```

```python
'''
Lets visualize global avarage traffic for every day 
'''

df = pd.read_csv('data/global_avarage_traffic.csv')

# Grupuj dane wedug dnia tygodnia
grouped_data = df.groupby('Dzien tygodnia')

# Twrz wykres dla kadego dnia tygodnia
for day, group in grouped_data:
    plt.plot(group['Godzina'], group['Nateenie'], label=day)

# Dodaj etykiety i legend
plt.title('Zmiany natenia pojazdw w cigu dnia')
plt.xlabel('Godzina')
plt.ylabel('Natenie')
plt.legend(title='Dzie tygodnia')

# Wywietl wykres
plt.show()
```
